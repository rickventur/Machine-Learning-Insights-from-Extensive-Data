```python
import pandas as pd
import matplotlib.pyplot as plt

clientes_df = pd.read_csv(r"C:\Users\Austin\Downloads\clientes.csv") # Importando a base de dados
display(clientes_df)
```

```python
# Verificando se temos valores vazios ou valores reconhecidos em formato errado.
display(clientes_df.info())
display(clientes_df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
# Vai transformar as colunas de texto em números, ex: profissionais vão sair de cientistas, professor, medico, etc para "0, 1, 2".
codificador = LabelEncoder()

# Só não aplicamos na coluna de score_credito que é o nosso objeto
for coluna in clientes_df.columns:
    if clientes_df[coluna].dtype == 'object' and coluna != 'score_credito':
        clientes_df[coluna] = codificador.fit_transform(clientes_df[coluna])

# Verificando se realmente todas as colunas foram modificadas 
for coluna in clientes_df.columns:
    if clientes_df[coluna].dtype == 'object' and coluna != 'score_credito':
        print[coluna]

display(clientes_df)
```

```python
# Escolhendos quais colunas vamos usa para treinar o modelo.
# Y é a coluna que queremos o modelo calculo.
# X vão ser todas as colunas que vamos usar para prever o score de credito, não vamos usar a coluna id_cliente porque ela é um número que não ajuda a previsão.

x = clientes_df.drop(['score_credito', 'id_cliente'], axis=1)
y = clientes_df['score_credito']

from sklearn.model_selection import train_test_split

# Separamos os dados em treino e teste. Treino vamos dar para os modelos aprender e teste vamos usar para ver se o modelo aprendendeu corretamente.
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvore = RandomForestClassifier() # Modelo arvore de decisão.
modelo_knn = KNeighborsClassifier() # Modelo do KNN.

# Treinando os modelos.
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Se o nosso modelo chutasse tudo em "standart", qual seria a acurácia do modelo?
contagem_scores = clientes_df['score_credito'].value_counts()
print(contagem_scores['Standard'] / sum(contagem_scores))
```

```python
from sklearn.metrics import accuracy_score

# Calculamos as previsões
previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

# Comparamos as previsões com o y_teste
# Esse score queremos os maior (maior acurácia, mas támbem tem que sair maior do que o chute de tudo Standard)
print(accuracy_score(y_teste, previsao_arvore))
print(accuracy_score(y_teste, previsao_knn))
```
